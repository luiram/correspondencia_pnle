## 1. Importar librerias

```python
import pandas as pd
from openpyxl import load_workbook
from num2words import num2words
import os  # Importar el módulo os
```

## 2. Carga de datos

```python
# Cargar la tabla de datos contratista

datos_df = 'BASE_DATOS_COMISIONISTAS V02.xlsx'
datos_df = pd.read_excel(datos_df,
                          sheet_name = 'BASE_DATOS',
                          header = 1,
                          usecols = 'A:CS'
                         )
```

```python
# número de filas y columnas
print(datos_df.shape)

# Identificación de columnas catagoricas y numéricas del set de datos
datos_df.info()
```

```python
# Ruta Plantilla Requerimiento Comisiones
planilla_requerimiento = 'Plantilla_Requerimiento_Comisiones.xlsx'
```

```python
# Crear la carpeta de salida si no existe
output_folder = 'Documentos_Equivalente'
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
```

## 3. Cruce de correspondencia

```python
for index, row in datos_df.iterrows():
    # Cargar la plantilla
    wb = load_workbook(planilla_requerimiento)
    ws = wb.active

    # Rellenar la plantilla con los datos del cliente
    ws['D3'] = f"{row['NUMERO REQUERIMIENTO']}"
    ws['D6'] = f"{row['NOMBRES']}"
    ws['I6'] = f"{row['CC']}"

    # Formatear la fecha de nacimiento como texto
    fecha_nacimiento = row['FECHA NACIMIENTO']
    if pd.notna(fecha_nacimiento):
        fecha_nacimiento_str = fecha_nacimiento.strftime('%d-%m-%Y')
    else:
        fecha_nacimiento_str = ''
    ws['L6'] = f"{fecha_nacimiento_str}"
    
    ws['D7'] = f"{row['CORREO']}"
    ws['J7'] = f"{row['CELULAR']}"

    # Formatear la fecha de inicio como texto
    fecha_inicio = row['FECHA INICIO']
    if pd.notna(fecha_inicio):
        fecha_inicio_str = fecha_inicio.strftime('%d-%m-%Y')
    else:
        fecha_inicio_str = ''
    ws['D14'] = f"{fecha_inicio_str}"

    # Formatear la fecha de final como texto
    fecha_fin = row['FECHA FINAL']
    if pd.notna(fecha_fin):
        fecha_fin_str = fecha_fin.strftime('%d-%m-%Y')
    else:
        fecha_fin_str = ''
    ws['I14'] = f"{fecha_fin_str}"
    
    ws['L14'] = f"{str(row['DIAS DE PERMANENCIA'])}"
    ws['J7'] = f"{row['CELULAR']}"
    ws['A26'] = f"{row['NOMBRES']}"
    ws['C26'] = f"{row['CC']}"
    ws['D26'] = f"{fecha_nacimiento_str}"
    ws['E26'] = f"{row['IDA ORIGEN']}"
    ws['F26'] = f"{row['IDA DESTINO']}"

    ### Formatear la fecha de ida como texto
    fecha_ida = row['IDA FECHA']
    if pd.notna(fecha_ida):
        fecha_ida_str = fecha_ida.strftime('%d-%m-%Y')
    else:
        fecha_ida_str = ''
    ws['G26'] = f"{fecha_ida_str}"
    
    ws['H26'] = f"{row['IDA HORA SIGERIDA SALIDA']}"
    ws['I26'] = f"{row['REGRESO ORIGEN']}"
    ws['J26'] = f"{row['REGRESO DESTINO']}"

    ### Formatear la fecha de regreso como texto
    fecha_regreso = row['REGRESO FECHA']
    if pd.notna(fecha_regreso):
        fecha_regreso_str = fecha_regreso.strftime('%d-%m-%Y')
    else:
        fecha_regreso_str = ''
    ws['K26'] = f"{fecha_regreso_str}"
    
    #ws['L26'] = f"{row['HORA SIGERIDA REGRESO']}"
    
    ws['A36'] = f"{row['CIUDAD RESIDENCIA']}"
    ws['B36'] = f"{fecha_inicio_str}"
    ws['F36'] = f"{fecha_inicio_str}"

    # Agregar el valor intermunicipal ida
    valor_inter_ida = row['Hacia Quibdó']
    if pd.notna(valor_inter_ida):
        ws['L36'] = float(valor_inter_ida)  # Convertir a float

    ws['B37'] = f"{fecha_fin_str}"
    ws['D37'] = f"{row['CIUDAD RESIDENCIA']}"
    ws['F37'] = f"{fecha_fin_str}"

    # Agregar el valor intermunicipal regreso
    valor_inter_regreso = row['Hacia ciudad de origen']
    if pd.notna(valor_inter_regreso):
        ws['L37'] = float(valor_inter_regreso)  # Convertir a float

    # Agregar el valor de gasto de desplazamiento
    valor_gasto_desplazamiento = row['GASTOS DE DESPLAZAMIENTO']
    if pd.notna(valor_gasto_desplazamiento):
        ws['K56'] = float(valor_gasto_desplazamiento)  # Convertir a float
    
    ws['E39'] = f"{row['Observaciones especiales para la solicitud']}"
    ws['B46'] = f"{row['NOMBRES']}"
    ws['F55'] = f"{str(row['DIAS DE PERMANENCIA'])}"
    
    
    # Agregar el valor estimado del tiquete como número
    valor_estimado_tiquete = row['VALOR ESTIMADO TIQUETE']
    if pd.notna(valor_estimado_tiquete):
        ws['K61'] = float(valor_estimado_tiquete)  # Convertir a float
    
    # Nombre del archivo de salida
    nombre_archivo = os.path.join(output_folder, f"{row['NUMERO REQUERIMIENTO']}_Comision_{row['NOMBRES']}.xlsx")
    
    # Guardar la nueva factura
    wb.save(nombre_archivo)

print("Comisión generadas con éxito.")
```
